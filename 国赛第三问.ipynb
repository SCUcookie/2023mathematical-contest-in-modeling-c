```python
import pandas as pd

# 读取Excel文件
df = pd.read_excel('问题3.xlsx')

# 选择单品编码、销售日期和销量列
selected_columns = ['单品编码', '销售日期', '销量']
df = df[selected_columns]

# 将销售日期列转换为日期类型
df['销售日期'] = pd.to_datetime(df['销售日期'], format='%Y/%m/%d')

# 筛选出在6/24到6/30之间的数据
start_date = pd.to_datetime('2023/6/24', format='%Y/%m/%d')
end_date = pd.to_datetime('2023/6/30', format='%Y/%m/%d')
filtered_df = df[(df['销售日期'] >= start_date) & (df['销售日期'] <= end_date)]

# 创建一个新的DataFrame，以单品编码为索引，日期为列
pivot_df = filtered_df.pivot_table(index='单品编码', columns='销售日期', values='销量', aggfunc='sum', fill_value=0)

# 重新排序列，以保证日期顺序正确
date_range = pd.date_range(start_date, end_date)
pivot_df = pivot_df[date_range]

# 显示结果
pivot_df
```

## 去掉这七天中3天以上销量为0或者7天中销量小于2.5的天数大于5天的单品

```python
# 定义函数来检查条件
def filter_single_item(item_row):
    # 统计销量为0或小于2.5的天数
    zero_sales_count = (item_row == 0).sum()
    below_2_5_sales_count = (item_row < 2.5).sum()
    
    # 如果3天以上销量为0或者7天中销量小于2.5的天数大于5天，则返回False，否则返回True
    return not (zero_sales_count >= 3 or below_2_5_sales_count > 5)

# 应用过滤函数到每一行（单品）上
filtered_items = pivot_df.apply(filter_single_item, axis=1)

# 通过索引获取满足条件的单品
selected_items = pivot_df[filtered_items]

# 显示结果
selected_items
```

## 最小二乘简略估计7/1销量

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 创建一个DataFrame来存储预测结果
predictions_df = pd.DataFrame(columns=['单品编码', '7/1预测销量'])

# 遍历每个单品
for item_code in selected_items.index:
    # 选择该单品的销量时间序列数据
    sales_data = selected_items.loc[item_code]
    
    # 创建自变量 X，表示日期
    X = np.arange(1, 8).reshape(-1, 1)
    
    # 创建因变量 y，表示销量
    y = sales_data.values.reshape(-1, 1)
    
    # 创建线性回归模型并拟合数据
    model = LinearRegression()
    model.fit(X, y)
    
    # 预测第8天（7/1）的销量
    forecast_7_1 = model.predict(np.array([[8]]))[0][0]
    
    # 将预测结果添加到DataFrame中
    predictions_df = predictions_df.append({'单品编码': item_code, '7/1预测销量': forecast_7_1}, ignore_index=True)

# 显示预测结果
predictions_df
```

```python
import pandas as pd

# 读取'2023.xlsx'中的销售数据
sales_data = pd.read_excel('2023.xlsx')

# 假设filtered_forecast包含了已筛选的单品列表
filtered_items = filtered_forecast.index.tolist()

# 筛选出在filtered_items中的单品
filtered_sales_data = sales_data[sales_data['单品编码'].isin(filtered_items)]

# 将销售日期列转换为日期类型
filtered_sales_data['销售日期'] = pd.to_datetime(filtered_sales_data['销售日期'], format='%Y/%m/%d')

# 创建一个日期范围，包含2023年的所有日期
date_range = pd.date_range(start='2023-01-01', end='2023-6-30')

# 创建一个DataFrame来存储每一天每一个单品的销售量和销售额
result_df = pd.DataFrame(index=date_range, columns=filtered_items)

# 将销售日期设置为索引
filtered_sales_data.set_index('销售日期', inplace=True)

# 计算每一天每一个单品的销售量和销售额
for date in date_range:
    for item in filtered_items:
        daily_sales = filtered_sales_data[(filtered_sales_data.index == date) & (filtered_sales_data['单品编码'] == item)]
        
        total_quantity = daily_sales['销量'].sum()  # 销售总量
        total_revenue = daily_sales['销售额'].sum()  # 销售总额
        
        result_df.loc[date, item] = total_quantity  # 将销售总量存储在对应的单品列中
        result_df.loc[date, f"{item}_销售额"] = total_revenue  # 将销售总额存储在对应的单品列中

# 填充缺失值为0
result_df.fillna(0, inplace=True)

# 显示结果
result_df
```
